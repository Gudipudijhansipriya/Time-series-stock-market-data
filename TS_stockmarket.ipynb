```python
import matplotlib.pylab as plt
import pandas as pd
import seaborn as sns
```

### Google stock data


```python
### 1 Google stock
g_stock=pd.read_csv("C:\\Users\\Lenovo\\OneDrive\\Desktop\\Timeseries_lab\\GOOG Stock data.csv") 
g_stock
```

```python
fig, ax = plt.subplots(2, 3, figsize=(14, 8))
fig.suptitle('\nDistribution\n', color='blue', fontsize=20)

# Title of each Distribution
ax[0,0].set_title('Open price distribution', color='blue')
ax[0,1].set_title('High price distribution', color='blue')
ax[0,2].set_title('Low price distribution', color='blue')
ax[1,0].set_title('Close price distribution', color='blue')
ax[1,1].set_title('Adjacent Close price distribution', color='blue')
ax[1,2].set_title('Volume price distribution', color='blue')

sns.kdeplot(ax=ax[0,0], x=g_stock.Open, shade=True, color='green')
sns.kdeplot(ax=ax[0,1], x=g_stock.High, shade=True, color='blue')
sns.kdeplot(ax=ax[0,2], x=g_stock.Low, shade=True, color='red')
sns.kdeplot(ax=ax[1,0], x=g_stock.Close, shade=True, color='magenta')
sns.kdeplot(ax=ax[1,1], x=g_stock.AdjClose, shade=True, color='purple')
sns.kdeplot(ax=ax[1,2], x=g_stock.Volume, shade=True, color='black')

plt.show()
```

* Conclusion:  We remark that *Open Price, High price, Low price, Close price* and 

* Adjacent price* have the same distribution and *Volume* distribution is different.
 
* Therefore the analysis will be based on Volume and the open price values

```python
plt.plot(g_stock.Date,g_stock.Open)
plt.rcParams["figure.figsize"] = [20,9]
plt.rcParams["figure.autolayout"] = True
plt.xlabel("Date")
plt.ylabel("Open")
plt.title("plot for date and open")
plt.show()
```

* Interpretation : There is exponential trend in the serie

```python
plt.plot(g_stock.Date,g_stock.High)
plt.rcParams["figure.figsize"] = [20,9]
plt.rcParams["figure.autolayout"] = True
plt.xlabel("Date")
plt.ylabel("High")
plt.title("plot fot date and high")
plt.show()
```

```python
plt.plot(g_stock.Date,g_stock.Low)
plt.rcParams["figure.figsize"] = [20,9]
plt.rcParams["figure.autolayout"] = True
plt.xlabel("Date")
plt.ylabel("Low")
plt.title("plot for date and low")
plt.show()
```

```python
plt.plot(g_stock.Date,g_stock.Close)
plt.rcParams["figure.figsize"] = [20,9]
plt.rcParams["figure.autolayout"] = True
plt.xlabel("Date")
plt.ylabel("Close")
plt.title("plot for date and close")
plt.show()
```

```python
plt.plot(g_stock.Date,g_stock.AdjClose)
plt.rcParams["figure.figsize"] = [20,9]
plt.rcParams["figure.autolayout"] = True
plt.xlabel("Date")
plt.ylabel("AdjClose")
plt.title("plot for date and adjclose")
plt.show()
```

```python
plt.plot(g_stock.Date,g_stock.Volume)
plt.rcParams["figure.figsize"] = [20,9]
plt.rcParams["figure.autolayout"] = True
plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("plot for date and volume")
plt.show()
```

* Interpretation: We can find a stationnarity and Unexpected event
